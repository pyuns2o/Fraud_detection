```python
from pandasql import sqldf
dfsql = lambda q: sqldf(q, globals())
```

```python
risk_df = dfsql('''select CLLT_FP_PRNO, count(CLLT_FP_PRNO) as risk from yes_fp group by CLLT_FP_PRNO having risk > 24 ORDER BY risk DESC''')

risk_df
```
